```python
import pymysql

db = pymysql.connect("localhost", 'root', '', 'cms')
cursor = db.cursor()



class Phonebook:
    def __init__(self):
        self.id=None
        self.name=None
        self.age=None
    def saveContact(self):
        q = "insert into cms3(name,age) values(%s,%s)"
        val = (self.name, self.age)
        cursor.execute(q, val)
        db.commit()
        message ="Data Store Successfully"
        print(message)
    def searchContact(self):
        q = f"select * from cms3 where id ={self.id}"
        cursor.execute(q)
        myresult = cursor.fetchall()#it return data in list.
        for x in myresult:
            print('id:',x[0])
            print('Name:',x[1])
            print('Age:',x[2])
   
        

    def allContact(self):
        q = f"select * from cms3"
        cursor.execute(q)
        myresult = cursor.fetchall()
        for x in myresult:
            print('id:',x[0] ,end=" ")
            print('Name:',x[1],end=" ")
            print('Age:',x[2],end=" ")
            print()
           



    def deleteContact(self):
        q = f"delete from cms3 where id ={self.id}"
        cursor.execute(q)
        db.commit()
        print(cursor.rowcount, "record(s) deleted")
while(True):

    print('1. for add \n2.for search\n3.for Delete\n4.showall\n 0 for exit')
    ch=input('enter choice:')
    if ch=='1':

        p= Phonebook()
        p.name=input('Enter Name:')
        p.age=input('Enter Age:')
        p.saveContact()
    elif ch=='3':

        p =Phonebook()
        p.id=int(input('enter id'))
        p.deleteContact()
        
    elif ch=='2':

        p = Phonebook()
        p.id=int(input('enter id'))
        p.searchContact()
    
    elif ch == '4':
        p =Phonebook()
        p.allContact()
    elif ch == '0':
        break
        
    else:
        print('invalid Choice')









```
